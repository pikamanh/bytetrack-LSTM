```python
import os

# Point this to the directory that contains your ReID image folders.
# The SportMOT class below also handles one nested dataset directory level.
os.environ["FASTREID_DATASETS"] = "/kaggle/input/datasets/dino1902/sportmot-reid"
print("FASTREID_DATASETS=", os.environ["FASTREID_DATASETS"])
```

```python
%%writefile /kaggle/working/fast-reid/fastreid/data/datasets/sportmot.py
# encoding: utf-8

import glob
import os.path as osp
import re

from .bases import ImageDataset
from ..datasets import DATASET_REGISTRY


@DATASET_REGISTRY.register()
class SportMOT(ImageDataset):
    """SportMOT ReID patches generated in Market1501-style folders."""

    dataset_name = "sportmot"

    def __init__(self, root="datasets", **kwargs):
        self.root = osp.abspath(osp.expanduser(root))
        self.dataset_dir = self._resolve_dataset_dir(self.root)
        self.train_dir = osp.join(self.dataset_dir, "bounding_box_train")
        self.query_dir = osp.join(self.dataset_dir, "query")
        self.gallery_dir = osp.join(self.dataset_dir, "bounding_box_test")

        required_files = [
            self.dataset_dir,
            self.train_dir,
            self.query_dir,
            self.gallery_dir,
        ]
        self.check_before_run(required_files)

        train = self.process_dir(self.train_dir)
        query = self.process_dir(self.query_dir, is_train=False)
        gallery = self.process_dir(self.gallery_dir, is_train=False)
        super(SportMOT, self).__init__(train, query, gallery, **kwargs)

    @staticmethod
    def _resolve_dataset_dir(root):
        candidates = [
            root,
            osp.join(root, "SportMOT-ReID"),
            osp.join(root, "sportmot-reid"),
            osp.join(root, "sportmot"),
        ]
        for candidate in candidates:
            if osp.isdir(osp.join(candidate, "bounding_box_train")):
                return candidate
        return candidates[0]

    def process_dir(self, dir_path, is_train=True):
        img_paths = glob.glob(osp.join(dir_path, "*.jpg")) + glob.glob(osp.join(dir_path, "*.png"))
        pattern = re.compile(r"([\-\d]+)_c(\d+)")
        data = []

        for img_path in sorted(img_paths):
            match = pattern.search(osp.basename(img_path))
            if match is None:
                continue
            pid, camid = map(int, match.groups())
            if pid <= 0:
                continue
            camid -= 1
            if is_train:
                pid = self.dataset_name + "_" + str(pid)
                camid = self.dataset_name + "_" + str(camid)
            data.append((img_path, pid, camid))

        if len(data) == 0:
            raise RuntimeError(
                "No valid ReID images found in {}. Expected names like 000001_c1*.jpg or .png".format(dir_path)
            )
        return data
```

```python
# Register SportMOT so DATASETS.NAMES can find it.
init_file = "/kaggle/working/fast-reid/fastreid/data/datasets/__init__.py"
with open(init_file, "r", encoding="utf-8") as f:
    content = f.read()
if "from .sportmot import SportMOT" not in content:
    with open(init_file, "a", encoding="utf-8") as f:
        f.write("\nfrom .sportmot import SportMOT\n")
```
